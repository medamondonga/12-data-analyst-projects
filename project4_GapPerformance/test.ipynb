```python
import numpy as np
from scipy.stats import shapiro, kstest
import seaborn as sns
```

## Shapiro-wilk test

```python
alpha = 0.05
```

```python
np.random.seed(15) #fix the value
```

```python
rolls = np.random.randint(1, 15, 30)
sns.histplot(rolls)
```

```python
stat, shapiro_p_value = shapiro(rolls)
print(shapiro_p_value)
```

## Kolmogorov-Smirnov test(KS)

```python
rolls = np.random.rand(100000)
sns.histplot(rolls)
```

```python
stat, ks_p_value = kstest(rolls, 'norm')
print(ks_p_value)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gnrer des donnes
np.random.seed(0)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(5, 2, 1000)
data3 = np.random.normal(-3, 0.5, 1000)

# Mettre dans un DataFrame pour seaborn
df = pd.DataFrame({
    "valeurs": np.concatenate([data1, data2, data3]),
    "distribution": (["N(0,1)"]*1000 + ["N(5,2)"]*1000 + ["N(-3,0.5)"]*1000)
})

# Plot avec seaborn
sns.displot(
    data=df,
    x="valeurs",
    col="distribution",       # un sous-graphe par distribution
    kde=True,                 # ajoute la courbe de densit ( Gauss)
    bins=30,                  # nombre de barres
    color="skyblue"           # couleur des histos
)

plt.show()
```
